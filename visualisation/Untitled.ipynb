```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cars.csv')
```

```python
plt.clf()
fig=plt.subplots(4,1,figsize=(40,20))

plt.subplot(2,2,1)
sns.scatterplot(x='mpg',y=' hp',data=df,hue=df[' brand'])
plt.xlabel('miles per gallon(mpg)')
plt.ylabel('horse power (hp)')
plt.title("mpg vs hp")


plt.subplot(2,2,2)
sns.scatterplot(x='mpg',y=' weightlbs',data=df,hue=df[' brand'])
plt.xlabel('miles per gallon(mpg)')
plt.ylabel('weightlbs')
plt.title("mpg vs weightlbs")

plt.subplot(2,2,3)
sns.scatterplot(x=' hp',y='mpg',data=df,hue=df[' brand'])
plt.xlabel('hp')
plt.ylabel('mpg')
plt.title("hp vs mpg")

plt.subplot(2,2,4)
sns.scatterplot(x= ' hp',y=' time-to-60',data=df,hue=df[' brand'])
plt.xlabel('hp')
plt.ylabel('time-to-60')
plt.title("hp vs time-to-60")

plt.show()
```

```python
plt.clf()
fig=plt.subplots(3,2,figsize=(40,20))

plt.subplot(2,2,1)
sns.lineplot(x='mpg',y=' hp',data=df,hue=df[' brand'])
plt.xlabel('miles per gallon(mpg)')
plt.ylabel('horse power (hp)')
plt.title("mpg vs hp")

plt.subplot(2,2,2)
sns.lineplot(x= ' hp',y=' time-to-60',data=df,hue=df[' brand'])
plt.xlabel('hp')
plt.ylabel('time-to-60')
plt.title("hp vs time-to-60")

plt.subplot(2,2,3)
sns.lineplot(x= 'mpg',y=' time-to-60',data=df,hue=df[' brand'])
plt.xlabel('mpg')
plt.ylabel('time-to-60')
plt.title("mpg vs time-to-60")

plt.subplot(2,2,4)
sns.lineplot(x=' hp',y=' weightlbs',data=df,hue=df[' brand'])
plt.xlabel('hp')
plt.ylabel('weightlbs')
plt.title("hp vs weightlbs")

plt.subplot(2,2,5)
sns.lineplot(x='mpg',y=' weightlbs',data=df,hue=df[' brand'])
plt.xlabel('mpg')
plt.ylabel('weightlbs')
plt.title("mpg vs weightlbs")

plt.subplot(2,2,6)
sns.lineplot(x='mpg',y=' cubicinches',data=df,hue=df[' brand'])
plt.xlabel('mpg')
plt.ylabel('cubicinches')
plt.title("mpg vs cubicinches")


plt.show()
```
