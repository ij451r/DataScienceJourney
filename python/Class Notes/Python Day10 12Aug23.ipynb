### JOINS/MERGE and CONCAT

```python
data_a = {'ID' : [1,2,3] , 'Value_A':['A1','A2','A3']}
data_b = {'ID' : [2,3,4] , 'Value_B':['B1','B2','B3']}
df_a = pd.DataFrame(data_a)
df_b = pd.DataFrame(data_b)
```

```python
# inner join
inner_join_result = pd.merge(df_a , df_b , on = 'ID' , how='inner')
```

```python
inner_join_result.head()
```

```python
# Outer 
outer_join_result = pd.merge(df_a,df_b, on='ID' , how = 'outer')
outer_join_result.head()
```

```python
left_join_result = pd.merge(df_a, df_b, on = 'ID' , how = 'left')
left_join_result.head()
```

```python
right_join_result = pd.merge(df_a, df_b, on = 'ID' , how = 'right')
right_join_result.head()
```

```python
data_a = {'ID' : [1,2,3] , 'Value':['A1','A2','A3']}
data_b = {'ID' : [2,3,4] , 'Value':['B1','B2','B3']}
df_a = pd.DataFrame(data_a)
df_b = pd.DataFrame(data_b)
```

```python
# Concat
df_a.head()
```

```python
df_b.head()
```

```python
concat_A_B_axis0 = pd.concat([df_a , df_b])
concat_A_B_axis0.head()
```

```python
concat_A_B_axis1 = pd.concat([df_a , df_b] , axis=1)
concat_A_B_axis1.head()
```
