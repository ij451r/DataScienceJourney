# Matplotlib

## Importing a library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# %matplotlib inline
```

## Create a line plot

```python
# define dataset
x = np.arange(0,10,0.1)
y = 2*x + 5

#plotting data points
plt.plot(x,y)
plt.show()
```

## Customizing the line plot

```python
# define dataset
x = np.arange(0,10,0.1)
y = 2*x + 5

#change a figure size
fig = plt.figure(figsize=(20,5))

plt.plot(x, y, linewidth = 2.0 , linestyle = ":", color = 'r' , alpha = 0.5 , marker = 'o')
plt.title("Line plot customization")
plt.xlabel("X_Axis")
plt.ylabel("Y_Axis")
plt.legend(['line1'] , loc  = 'best')
plt.grid(True)
plt.show()
```

 ## Subplots
 
- In this particular example we are creating 2*2 grid 
- First argument specifies total no of rows
- 2nd arugument specifies total no of columns
- 3rd argument specifies the index of the current subplot with in that grid

```python
# define dataset
x = np.arange(0,10,0.1)
y1 = 2*x + 5
y2 = 3*x + 10
y3 = 4*x + 15
y4 = 5*x + 20
#crrate 2 or more plots in a figure
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title('Graph1')

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title('Graph2')

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('Graph3')

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title('Graph4')

plt.show()

```

## Create a bar plot

```python
# creating a dataset

data = {'apples' : 20 , 'Mangoes' :15 , 'Lemon':30 , 'Oranges':10}
names = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize=(10,5))
plt.bar(names,values)
plt.show()
```

```python
# Bar plot with customization
# creating a dataset

data = {'apples' : 20 , 'Mangoes' :15 , 'Lemon':30 , 'Oranges':10}
names = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize=(10,5))
plt.bar(names,values,color = ['red','yellow','yellow','orange'])
plt.show()
```

## horizontal bar plot along with customization

```python
# creating a dataset

data = {'apples' : 20 , 'Mangoes' :15 , 'Lemon':30 , 'Oranges':10}
names = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize=(10,5))
plt.barh(names,values,color = "orange")
plt.title("Bar graph Demo")
plt.xlabel("Quantity")
plt.ylabel("Fruits")
plt.show()
```

## Create Scatter plot

```python
a = [10,20,30,40,50,60,70,80]
b = [5,3,2,5,6,1,4,2]
x = [1,2,3,4,3,2,1,2]

# create a plot

plt.scatter(a,b , c ='g' ,  s = 200 , edgecolors='y' , marker = 'o' , alpha = 0.5)
plt.scatter(a,x , c ='r' , s = 400 , edgecolors='b' , marker = '4' , alpha = 1)
plt.legend(['b','x'] , loc = 'best')
plt.title("Scatter plt Demo")
plt.xlabel("X_Axis")
plt.ylabel("Y_Axis")
plt.grid(True)
plt.savefig('Scatter PLot.png')
plt.show()
```

## Create Histograms

## Create a Box Plot

```python
# dataset
total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,20,10,5,10,20,30,40,20,45]
data = list([total,order,discount])
#plot the data
plt.boxplot(data)
plt.show()
```

## Create a Pie Chart

```python
#dataset
label = ['Dog','Cat','Wolf','Lion']
sizes = [50,45,60,80]

plt.pie(sizes , labels=label , autopct='%1.1f%%' , shadow = True , explode = (0,0.1,0.1,0) , startangle=90)
plt.show()
```

## Create a Doughnut Chart

```python
name = ['GrpA','GrpB','GrpC']
size = [20,30,50]

size_centre = [5]
a,b,c = [plt.cm.Blues , plt.cm.Reds , plt.cm.Greens]

pie1 = plt.pie(size , labels = name , radius = 1.5 , colors=[a(0.5) , b(1.5) , c(0.5)])
pie2 = plt.pie(size_centre,radius=1,colors='w')
plt.show()
```

# Seaborn

## Import Library

```python
import seaborn as sns
```

## Import Dataset from seaborn

```python
flights = sns.load_dataset("flights")
flights.head()
```

## Create Lineplot

```python
sns.lineplot(data = flights , x = "year" , y = "passengers" , color = "r")
plt.title("Number of passengers by year")
plt.xlabel("Years")
plt.ylabel("passengers")
plt.show()

```

```python
flights.dtypes
```

```python
flights.dtypes
```

```python
flights['year'] = flights['year'].astype('category')
```

```python
flights.dtypes
```

```python
flights_grouped = flights.groupby("year").sum(numeric_only=True)
```

```python
flights_grouped = flights[['year','passengers']].groupby("year").sum()
flights_grouped.head()
```

```python
flights_grouped = flights_grouped.reset_index()
flights_grouped.head()
```

```python
sns.barplot(data=flights_grouped,x='year',y='passengers' )
```

```python
flighta_wide = flights.pivot(index="year" , columns="month" , values="passengers")
flighta_wide.head()
```

```python
sns.lineplot(data=flighta_wide["Jun"])
plt.title("Flights in June")
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(data=flights , x = "year" , y = "passengers" , hue="month")
plt.title("Number of passengers by year on monthly basis")
plt.show()
```
