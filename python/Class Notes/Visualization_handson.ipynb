1. Create scatter plots for the following data, make sure all the plots appear in the same
plane with x-labels and y-labels evenly spaced.
- Create scatter plots with between (mpg-hp, mpg-weight), and (hp-mpg,
hp-time-to-60) with respect to each brand.


```python
fig=plt.subplots(2,2,figsize=(20,10))

plt.subplot(2,2,1)
sns.scatterplot(x='mpg',y=' hp',data=df,hue=df[' brand'])
plt.xlabel('miles per gallon(mpg)')
plt.ylabel('horse power (hp)')
plt.title("mpg vs hp")


plt.subplot(2,2,2)
sns.scatterplot(x='mpg',y=' weightlbs',data=df,hue=df[' brand'])
plt.xlabel('miles per gallon(mpg)')
plt.ylabel('weightlbs')
plt.title("mpg vs weightlbs")

plt.subplot(2,2,3)
sns.scatterplot(x=' hp',y='mpg',data=df,hue=df[' brand'])
plt.xlabel('hp')
plt.ylabel('mpg')
plt.title("hp vs mpg")

plt.subplot(2,2,4)
sns.scatterplot(x= ' hp',y=' time-to-60',data=df,hue=df[' brand'])
plt.xlabel('hp')
plt.ylabel('time-to-60')
plt.title("hp vs time-to-60")

plt.show()
```

2. Create bar plots for the following, make sure all the plots appear in the same plane with
x-labels and y-labels evenly spaced.
- Create a bar plot that shows the visual representation of hp, mpg, weight,
time-to-60 with respect to the number of cylinders for each of the brands.
- Create a bar plot that shows the visual representation of hp, mpg, weight and
time-to-60 with respect to the years, for each brand.


```python
fig = plt.subplots(2,2, figsize = (20,10))

plt.subplot(2,2,1)
sns.barplot(x = ' cylinders', y = ' hp', data = df, hue = df[' brand'] )
plt.xlabel('cylinders')
plt.ylabel('hp')
plt.title('hp vs cylinders')

plt.subplot(2,2,2)
sns.barplot(x = ' cylinders', y = 'mpg', data = df, hue = df[' brand'] )
plt.xlabel('cylinders')
plt.ylabel('mpg')
plt.title('mpg vs cylinders')

plt.subplot(2,2,3)
sns.barplot(x = ' cylinders', y = ' weightlbs', data = df, hue = df[' brand'] )
plt.xlabel('cylinders')
plt.ylabel('weightlbs')
plt.title('weightlbs vs cylinders')

plt.subplot(2,2,4)
sns.barplot(x = ' cylinders', y = ' time-to-60', data = df, hue = df[' brand'] )
plt.xlabel('cylinders')
plt.ylabel('time-to-60')
plt.title('time-to-60 vs cylinders')

plt.show()
```

```python
fig = plt.subplots(2,2, figsize = (20,10))

plt.subplot(2,2,1)
sns.barplot(x = ' year', y = ' hp', data = df, hue = df[' brand'] )
plt.xlabel('year')
plt.ylabel('hp')
plt.title('hp vs year')

plt.subplot(2,2,2)
sns.barplot(x = ' year', y = 'mpg', data = df, hue = df[' brand'] )
plt.xlabel('year')
plt.ylabel('mpg')
plt.title('mpg vs year')

plt.subplot(2,2,3)
sns.barplot(x = ' year', y = ' weightlbs', data = df, hue = df[' brand'] )
plt.xlabel('year')
plt.ylabel('weightlbs')
plt.title('weightlbs vs year')


plt.subplot(2,2,4)
sns.barplot(x = ' year', y = ' time-to-60', data = df, hue = df[' brand'] )
plt.xlabel('years')
plt.ylabel('time-to-60')
plt.title('time-to-60 vs years')

plt.show()
```
